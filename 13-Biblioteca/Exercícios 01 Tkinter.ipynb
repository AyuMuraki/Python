### Execute as etapas abaixo para criar uma programa que calcule o IMC:

1. Importe a biblioteca **tkinter**.
1. Crie uma label com o título **Cálculo de IMC**.
1. Crie uma label com o texto **Digite o seu Peso**.
1. Utilizando o método Entry, crie uma entrada de texto para receber o peso do usuário.
1. Crie uma label com o texto **Digite a sua Altura**.
1. Utilizando o método Entry, crie uma entrada de texto para receber a altura do usuário.
1. Crie uma função chamada **calcular** para calcular o IMC do usuário:
    * A função deve ser criada no início do programa
    * A formúla do IMC é: **Peso / (Altura*Altura)**
1. Utilizando o método Button, crie um botão com os parâmentros abaixo:
    * Insira o texto **Calcular** para o parâmetro **text**.
    * Insira a função **calcular** para o parâmetro **command**.
1. Crie uma label sem texto que receberá o resultado do cálculo do IMC quando o botão for acionado.

```python
import tkinter  # Importa a biblioteca tkinter para criar interfaces gráficas

# Inicializa a aplicação
app = tkinter.Tk()

# Função

def somar ():
    #print('Botão precisonado')


    peso =  int (entrada_num1.get())
    altura = int (entrada_num2.get())

    total = peso / (altura * altura)

    cx_txt_resultado.configure(text = total)

# Define as dimensões da janela do aplicativo (600x400 pixels)
app.geometry('250x200')

# Cria um rótulo (Label) com o texto 'Somar Valores', definindo a fonte, tamanho e estilo em negrito
cx_txt = tkinter.Label(app, text='Cálculo de IMC', font=('Georgia', 14, 'bold'), fg = '#fc0388' , pady= 10)

# Posiciona o rótulo na linha 0 e coluna 0, ocupando 2 colunas (centralizando)
cx_txt.grid(column=0, row=0, columnspan=2)

# Cria um rótulo para o Número 1
cx_txt_num1 = tkinter.Label(app, text='Digite o seu Peso:')

# Posiciona o rótulo do Número 1 na linha 1 e coluna 0
cx_txt_num1.grid(column=0, row=1)

# Cria um campo de entrada de texto para o usuário inserir o Número 1
entrada_num1 = tkinter.Entry(app)  # Entrada de dados

# Posiciona a entrada de texto na linha 1 e coluna 1 (ao lado do rótulo Número 1)
entrada_num1.grid(column=1, row=1)

# Cria um rótulo para o Número 2
cx_txt_num2 = tkinter.Label(app, text='Digite a sua Altura:')

# Posiciona o rótulo do Número 2 na linha 2 e coluna 0
cx_txt_num2.grid(column=0, row=2)

# Cria um campo de entrada de texto para o usuário inserir o Número 2
entrada_num2 = tkinter.Entry(app)

# Posiciona a entrada de texto na linha 2 e coluna 1 (ao lado do rótulo Número 2)
entrada_num2.grid(column=1, row=2, pady= 10, padx = 15)

# Cria um botão com o texto 'Calcular', sem uma função definida no momento
botao = tkinter.Button(app, text='Calcular', command=somar , width=15)

# Posiciona o botão na linha 3, ocupando 2 colunas (centralizando)
botao.grid(column=0, row=3, columnspan=2)

# Resultado 
cx_txt_resultado = tkinter.Label(app, text= ' total')

cx_txt_resultado.grid(column=0,row=4, columnspan=2, pady = 10)

# Inicia o loop principal da aplicação, fazendo com que a janela permaneça aberta
app.mainloop()
```
